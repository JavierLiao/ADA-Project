### 3.1 Missing data

```python
missing = train.isna().sum().sort_values(ascending = False)
missing
```

```python
print(f'Totle number of missing data:{train.isna().sum().sum():d}')
```

```python
## the plot

font = {'fontsize':12, 'fontweight':'bold'}
font2 = {'fontsize':10, 'fontweight':'bold'}
#palette = ["#682F2F", "#9E726F", "#D6B2B1", "#B9C0C9", "#9F8A78", "#F3AB60"]
fig, ax = plt.subplots(1, 1, figsize = (8, 10))
sns.barplot(x = missing.values, y = missing.index, palette='viridis')
plt.xlabel('Number', fontdict=font)
plt.title('The Number of Missing Data for Each Column', fontdict=font);
```

```python
def drop_missing(df):
    ## drop rows with missing rate > 0.15
    missing_rate_row = df.T.isna().sum() / df.shape[1]
    df_cleaned_row = df[missing_rate_row <= 1.5]

    ## drop columns with missing rate > 0.1
    missing_rate_col = df.isna().sum() / df.shape[0]
    messy_columns = df.columns[missing_rate_col > .1]
    df_cleaned = df_cleaned_row.drop(columns=messy_columns)

    ## reindex
    df_cleaned.index = np.arange(df_cleaned.shape[0], dtype = int)

    ## backward propagation fillna
    df_cleaned.fillna(method = 'ffill', inplace = True)

    ## datatype modify
    ## as some of the features' datatype does not accod to common sense!
    modify_to_float = ['Client_Income','Credit_Amount', 'Loan_Annuity','Population_Region_Relative',
                       'Age_Days', 'Employed_Days', 'Registration_Days', 'ID_Days','Score_Source_2']
    modify_to_int = ['Car_Owned', 'Bike_Owned', 'Active_Loan', 'House_Own']

    for col in modify_to_float:
        df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')
    df_cleaned[modify_to_int] = df_cleaned[modify_to_int].astype(int)
    df_cleaned.dropna(inplace= True)
    
    return df_cleaned

train_cleaned = drop_missing(train)
assert train_cleaned.isna().sum().sum() == 0 #non-null
```

### 4.1 Outliers

```python
## Screen outliers
num_col_new= num_col[1:-1]

def screen_outlier(df):
    df_screened = df.copy()
    for col in num_col_new:
        Q1 = np.percentile(df_screened[col], 25)
        Q3 = np.percentile(df_screened[col], 75)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR
        df_screened = df_screened[(df_screened[col] >= Q1 - outlier_step) & 
                                  (df_screened[col] <= Q3 + outlier_step)]
    return df_screened

train_screened = screen_outlier(train_cleaned)
print(f'Totoal number of rows with outliers: {train_cleaned.shape[0] - train_screened.shape[0]}')
```

```python
## Compare between before screening and after screening

fig, ax = plt.subplots(1,2, figsize = (12, 6))

sns.distplot(x = train_cleaned.Credit_Amount, bins=100, ax=ax[0]);
ax[0].set_title('Before Screening', fontdict=font)
ax[0].axis('off');

sns.distplot(x = train_screened.Credit_Amount, bins=100, ax=ax[1]);
ax[1].set_title('After Screening', fontdict=font);
ax[1].axis('off');
```

### 4.2 Skewing

```python
# Deal with skewing by sqrt()

def skew_sqrt(df):
    df_skewed = df.copy()
    for col in num_col_new:
        df_skewed[col] = df_skewed[col].apply(np.sqrt)
    return df_skewed

train_skewed = skew_sqrt(train_screened)
```

```python
## Compare between before screening and after screening

fig, ax = plt.subplots(1,3, figsize = (12, 4))

sns.distplot(x = train_cleaned.Credit_Amount, bins=100, ax=ax[0]);
ax[0].set_title('Origin', fontdict=font)
ax[0].axis('off');

sns.distplot(x = train_screened.Credit_Amount, bins=100, ax=ax[1]);
ax[1].set_title('After Screening', fontdict=font);
ax[1].axis('off');

sns.distplot(x = train_skewed.Credit_Amount, bins=100, ax=ax[2]);
ax[2].set_title('After Screening and Sqrt', fontdict=font)
ax[2].axis('off');
```

### 4.4 Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train_num_scaled = scaler.fit_transform(train_skewed[num_col_new])

train_num = pd.DataFrame(train_num_scaled, columns=num_col_new)
train_num.head()
```

### 4.4 Feature Engineering

```python
## one-hot encoding

train_cat = pd.get_dummies(train_skewed[cat_col])
train_cat.index = np.arange(train_cat.shape[0], dtype = int)

train_cat.head()
```

### 4.5 Final Output

```python
## Combine num and cat

train_final = pd.concat([train_num, train_cat], axis = 1)
train_final.head()
```

```python
## to csv
train_final.to_csv('../output/train_final.csv')
```

##  Process Test

```python
## process Test

test_cleaned = drop_missing(test)
test_screened = screen_outlier(test_cleaned)
test_skewed = skew_sqrt(test_screened)

test_num_scaled = scaler.fit_transform(test_skewed[num_col_new])
test_num = pd.DataFrame(test_num_scaled, columns=num_col_new)

test_cat = pd.get_dummies(test_skewed[cat_col])
test_cat.index = np.arange(test_cat.shape[0], dtype = int)

test_final = pd.concat([test_num, test_cat], axis = 1)
test_final.head()
```
